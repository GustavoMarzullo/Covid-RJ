```python
def comparar(a,b,porcemmil=False):
    '''Compara o estado a com o estado b. Plota um gr√°fico de casos acumulados.'''
    nao_eh_cidade=df['codmun'].isnull() #True para todas as linhas com c√≥digo municipal nulo
    
    eh_noa=df['estado']==str(a) # pega tudo do estado a
    casos_a=df[eh_noa & nao_eh_cidade] #faz a interse√ß√£o dos dois
    
    eh_nob=df['estado']==str(b) # pega tudo do estado b
    casos_b=df[eh_nob & nao_eh_cidade] #faz a interse√ß√£o dos dois
    
    #plota o gr√°fico
    if porcemmil==False:
        casos_a['casosAcumulado'].plot(color='black')
        casos_b['casosAcumulado'].plot(color='red')
        plt.legend([str(a),str(b)])
        plt.xlabel('Data')
        plt.ylabel('Casos Acumulados')
        plt.title('Casos Acumulados '+str(a)+' x '+str(b))
        plt.show()
    else:
        casos_aporcemmil=1e4*casos_a['casosAcumulado']/casos_a['populacaoTCU2019'].astype(int) #convertendo para casos por milh√£o
        casos_bporcemmil=1e4*casos_b['casosAcumulado']/casos_b['populacaoTCU2019'].astype(int)
        casos_aporcemmil.plot(color='black')
        casos_bporcemmil.plot(color='red')
        plt.legend([str(a),str(b)])
        plt.xlabel('Data')
        plt.ylabel('Casos Acumulados por 100K')
        plt.title('Casos Acumulados por 100K '+str(a)+' x '+str(b))
        plt.show()
```

```python
comparar('RJ','SP')
```

```python
comparar('RS','RJ',True)
```

```python
comparar('RJ','BA')
```

```python
pd.set_option('mode.chained_assignment', None)
def lista_casos(data,df=df):
    '''Data do tipo str: yyyy-mm-dd'''
    estados=df[(df['codmun'].isnull()) &(df['estado'].isnull()==False) & (df.index==data)]
    estados['Casos por cem mil']=round(1e4*estados['casosAcumulado']/estados['populacaoTCU2019'].astype(int),1)
    estados.set_index('estado',inplace=True)
    return estados[['Casos por cem mil']].sort_values('Casos por cem mil',ascending=False)
```

```python
lista_casos('2020-10-05')
```
